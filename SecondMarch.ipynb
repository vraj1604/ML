```python
import pandas as pd
```

```python
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('/census_data.csv')
data 
```

```python
sns.catplot(x='age',y='marital_status',kind='box',hue='gender',data=data)
```

```python
sns.catplot(x='age',y='marital_status',kind='bar',hue='gender',data=data)
```

```python
sns.relplot(x='capital_loss',y='capital_gain',hue='marital_status',data=data)
```

```python
sns.catplot(x='age',kind='count',hue='gender',data=data)
```

```python
sns.catplot(x='age',y='marital_status',kind='violin',hue='gender',data=data)
```
