```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Question 1 - Load data set using pandas function and print last 5 records
dimdf = pd.read_csv('diamonds.csv')
dimdf.head()
```

```python
# Question 2 - Draw the bar chart of costliest and cheapest diamonds
dprice = dimdf['price']
pmax = max(dprice)
pmin = min(dprice)
print(pmin,pmax)
```

```python
sorted_data = dimdf.sort_values('price', ascending=False)

# Select the costliest and cheapest diamonds
top_diamonds = sorted_data.head(1)
bottom_diamonds = sorted_data.tail(1)

# Concatenate the top and bottom diamonds into a single DataFrame
selected_data = pd.concat([top_diamonds, bottom_diamonds])

# Create a bar chart of the selected diamonds
selected_data.plot(x='carat', y='price', kind='bar', figsize=(8, 6), grid=True)

# Set the chart title and axis labels
plt.title('Costliest and Cheapest Diamonds')
plt.xlabel('Carat')
plt.ylabel('Price')

# Display the chart
plt.show()
```

```python
# Question 3 - Draw the chart for Price vs Carat of Diamonds
xprice = dimdf['price']
ycarat = dimdf['carat']
plt.plot(xprice,ycarat)
plt.show()
```
