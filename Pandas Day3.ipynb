```python
mdf=pd.read_csv('Movie-Data.csv')
mdf
```

```python
mdf['Year']==2012
```

```python
#Print Data whose year is 2012 without making year AS INDEX
data=mdf[mdf['Year']==2012]
data
```

```python
#Print movies directed by Ridley Scott
data2=mdf[mdf['Director']=='Ridley Scott']
data2

```

```python
#Print Data whose year is after 2012 without making year AS INDEX
data=mdf[mdf['Year']>=2012]
data
```

```python
#Print Data whose year is between 2012 to 2018 without making year AS INDEX
data=mdf[(mdf['Year']>=2012) & (mdf['Year']<=2018) ]
data
```

```python
#Print Movies that were released between 2005-2010 , have rating above 8 but made below revenue percentage of 25 
data3=mdf[(mdf['Year']>=2005) & (mdf['Year']<=2010) & (mdf['Rating']>=8.0) & (mdf['Revenue (Millions)']<=25)]
data3
```
